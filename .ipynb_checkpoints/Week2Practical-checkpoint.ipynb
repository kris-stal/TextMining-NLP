### Question 3. Implementing Capture Groups in Regex
Create a regex and use the function `re.findall()` to find words that repeat twice consecutively. For example, find "run run" at the sentence "I will run run to the store". Explain why `re.findall()` result is not the entire repetition.

````python
# develop here

text = "I will run run to the store"
pattern = r'\b(\w+)\s+\1\b'
# \b is a word boundary
# (\w*) captures one or more word characters in the group
# \s+ matches one or more whitespace characters
# \1 references the first group (same word)



result = re.findall(pattern, text)
print(result)  # Output: ['run']

# Reason for re.findall() not outputting the entire repetition is that the pattern only captures the groups (\w*), then it only outputs
# the contents of the first captured group, not the entire match through the pattern.
````

# Language Model (N-Gram) - Warm up

Try the code below for create bigrams

````python
import nltk
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

nltk.download("punkt")

text = [" I am Sam  ", " Sam I am  ", " I do not like green eggs and ham  "]
bigram = []
for line in text: # for every line
    token = word_tokenize(line) # one token is one line
    bigram = bigram + list( # the bigram is the empty list + this new list:
        ngrams( # list called ngrams
            token, # with one token
            2, # for every two lines in the text
            pad_right=True,
            pad_left=True,
            left_pad_symbol="<s>",
            right_pad_symbol="</s>",
        )
    )
word_fd = FreqDist(bigram) # then creating a frequency distribution for this bigram
print(word_fd) # print the freqdist
print(word_fd.most_common()) # freqdist by most common
````

Try the code below for bigram probability calculation

````python
import nltk
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

nltk.download("punkt")
text = ["I am Sam", "Sam I am", "I do not like green eggs and ham"]

bigram_list = []
unigram_list = []

for line in text: # for every line
    tokens = word_tokenize(line) # one token is one line
    unigram_list.extend(["<s>"] + tokens + ["</s>"]) # the unigram list is every token surrounded by whitespace
    bigram_list.extend( # the bigram list is
        ngrams(
            tokens, # every token
            2, # for every 2 lines, so bigram
            pad_left=True,# enable padding for left and right
            pad_right=True,
            left_pad_symbol="<s>",
            right_pad_symbol="</s>", # set padding symbol as whitespace
        )
    )

unigram_counts = FreqDist(unigram_list)  # {'<s>': 3, 'I': 3, ...
bigram_counts = FreqDist(bigram_list)  # {('<s>', 'I'): 2, ('I', 'am'): 2, ...


def bigram_mle_prob(w1, w2, bigram_counts, unigram_counts): # maximum likelihood estimation for the bigram and unigram
    return bigram_counts.get((w1, w2), 0) / unigram_counts.get(w1, 1) # return the MLE


print("Bigram MLE Probabilities:\n")
for bigram, count in bigram_counts.items(): # for every token in bigrams
    w1, w2 = bigram
    prob = bigram_mle_prob(w1, w2, bigram_counts, unigram_counts)
    print(f"P({w2} | {w1}) = {prob:.4f}")
````

````python
import nltk
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
nltk.download("punkt", quiet=True)
text = [
"I am Tom"
"Tom I am"
"I do like watch videos on Youtube and watch films in cinema"
]
bigram = []
for line in text:
    token = nltk.word_tokenize(line)
    bigram = bigram + list(
    ngrams(
    token,
    2,
    pad_right=True,
    pad_left=True,
    left_pad_symbol="<s>",
    right_pad_symbol="</s>",
)
)
word_fd = FreqDist(bigram)
print(word_fd)
print(word_fd.most_common())
````

### Question 5. Generating Trigrams

Generate **trigrams** using the **Wall Street Journal corpus** (`nltk.corpus.treebank.words()`) and print the 20 most common with their occurrences.

````python
# develop here
myWords = nltk.corpus.treebank.words()

trigrams = []
for w in myWords: # for every sentance
    trigrams.extend(
        ngrams( # create trigrams with the sentance
            w, # with one sentance
            3, # for every three lines in the text, so trigram
            pad_right=True,
            pad_left=True,
            left_pad_symbol="<s>",
            right_pad_symbol="</s>",
        )
    )
word_fd = FreqDist(trigrams) # then creating a frequency distribution for this bigram
print(word_fd) # print the freqdist
print(word_fd.most_common(20)) # freqdist by most common
````

### Question 6. Computing Bigram Probabilities

Using the bigrams from the **WSJ corpus** `nltk.corpus.treebank.words()`), compute the **MSE probability** for bigrams ("This", "is") and ("the", "company").

````python
# develop here

myWords = nltk.corpus.treebank.words()

bigram_list = []
unigram_list = []

unigram_list = (["<s>"] + myWords + ["</s>"]) # the unigram list is every token (character) surrounded by whitespace
bigram_list = list(ngrams(unigram_list, 2)) # bigram which is just 2 of every item in unigram list

unigram_counts = FreqDist(unigram_list)  # {'<s>': 3, 'I': 3, ...
bigram_counts = FreqDist(bigram_list)  # {('<s>', 'I'): 2, ('I', 'am'): 2, ...


def bigram_mle_prob(w1, w2, bigram_counts, unigram_counts): # maximum likelihood estimation for the bigram and unigram
    return bigram_counts.get((w1, w2), 0) / unigram_counts.get(w1, 1) # return the MLE


print("Bigram MLE Probabilities:\n")
print(f"P(This | is) = {bigram_mle_prob('is', 'This', bigram_counts, unigram_counts):.4f}")
print(f"P(the | company) = {bigram_mle_prob('the', 'company', bigram_counts, unigram_counts):.4f}")

print(f"Occurrences of 'is': {unigram_counts['is']}")
print(f"Occurrences of ('is', 'This'): {bigram_counts[('is', 'This')]}")
print(f"Occurrences of 'the': {unigram_counts['the']}")
print(f"Occurrences of ('the', 'company'): {bigram_counts[('the', 'company')]}")
````

### Question 7. Applying Add-1 Smoothing
Using news category from **Brown corpus** (`nltk.corpus.brown.words(categories="news")`), calculate **Add-1 Smoothing** probability for `("the", "president")` and `(United", "States")` bigrams. Tip: Modify the code example for **MSE calculation** from the previous question.

````python
# develop here
myNews = nltk.corpus.brown.words(categories="news")
vocabSize = len(set(myNews))

bigram_list = []
unigram_list = []

unigram_list = (["<s>"] + myNews + ["</s>"]) # the unigram list is every token (character) surrounded by whitespace
bigram_list = list(ngrams(unigram_list, 2)) # bigram which is just 2 of every item in unigram list

unigram_counts = FreqDist(unigram_list)  # 
bigram_counts = FreqDist(bigram_list)  # 

def bigram_add1_smoothing(w1, w2, bigram_counts, unigram_counts, V):
    return(bigram_counts.get((w1, w2), 0) + 1) / (unigram_counts.get(w1, 0) + V)

print(f"P(president | the) = {bigram_add1_smoothing('president', 'the', bigram_counts, unigram_counts, vocabSize):.4f}")
print(f"P(United | States) = {bigram_add1_smoothing('United', 'States', bigram_counts, unigram_counts, vocabSize):.4f}")

````


### Question 8. Perplexity Calculation for a Sentence

Using the news category from **Brown corpus** (`nltk.corpus.brown.words(categories="news")`), implement a function to compute **perplexity** and test the sentence `"The president of the United States"`

````python
import math

# develop here
myNews = nltk.corpus.brown.words(categories="news")
vocabSize = len(set(myNews))

bigram_list = []
unigram_list = []

unigram_list = (["<s>"] + myNews + ["</s>"]) # the unigram list is every token (character) surrounded by whitespace
bigram_list = list(ngrams(unigram_list, 2)) # bigram which is just 2 of every item in unigram list

unigram_counts = FreqDist(unigram_list)  # 
bigram_counts = FreqDist(bigram_list)  # 

def bigram_add1_smoothing(w1, w2, bigram_counts, unigram_counts, V):
    return(bigram_counts.get((w1, w2), 0) + 1) / (unigram_counts.get(w1, 0) + V)

def calculate_perplexity(sentance, bigram_counts, unigram_counts, V):
    words = ["<s>"] + sentance.split() + ["<s>"] # sentance boundary
    sentance_bigrams = list(ngrams(words, 2)) # bigrams

    # Compute the sentence probability using the smoothed bigram probabilities
    log_prob_sum = 0
    for w1, w2 in sentance_bigrams:
        prob = bigram_add1_smoothing(w1, w2, bigram_counts, unigram_counts, V)
        log_prob_sum += math.log(prob)  # Take log to avoid underflow

    # Compute perplexity
        perplexity = math.exp(-log_prob_sum / len(sentance_bigrams))
    return(perplexity)
    

# Test Sentence
test_sentence = "The president of the United States"
perplexity = calculate_perplexity(test_sentence, bigram_counts, unigram_counts, vocabSize)

print(f"P(president | the) = {bigram_add1_smoothing('president', 'the', bigram_counts, unigram_counts, vocabSize):.4f}")
print(f"P(United | States) = {bigram_add1_smoothing('United', 'States', bigram_counts, unigram_counts, vocabSize):.4f}")
````
